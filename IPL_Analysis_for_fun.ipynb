```
def player_avg(player_name):
  player_data = merged_2020_data[(merged_2020_data['batsman'] == player_name)]

  player_runs = player_data['batsman_runs'].sum()
  player_data2 = merged_2020_data[(merged_2020_data['batsman'] == player_name) | (merged_2020_data['non_striker'] == player_name)]

  player_innings = len(player_data2['id'].unique())

  player_out_data = merged_2020_data[(merged_2020_data['player_dismissed'] == player_name)]
  out_total = len(player_out_data[player_out_data['dismissal_kind'] != 'retired hurt'])

  total_not_outs = player_innings - out_total

  
  if out_total == 0:
    return player_runs,player_runs,player_innings,total_not_outs
  else:
    return player_runs/out_total,player_runs,player_innings,total_not_outs
```

```
player_avg('SK Raina')
```

```
#import tqdm
from tqdm import tqdm
player_dict = {}
for each in tqdm(merged_2020_data['batsman'].unique()):
  avg , runs , inn , nos = player_avg(each)
  player_dict[each] = [inn, runs,nos, avg]
```

```
avg_details = pd.DataFrame(player_dict).transpose()
avg_details = avg_details.rename(columns ={0:'innings',1:'runs',2:'NOs',3:'average'})


avg_details[['innings', 'runs', 'NOs']] = avg_details[['innings', 'runs', 'NOs']].apply(np.int64)

avg_details = avg_details.reset_index()
avg_details.columns = ['batsman','no_of_innings','batsman_runs','not_out','avg_data']

avg_details = avg_details[avg_details['no_of_innings'] > 5].sort_values(['avg_data','batsman'] ,ascending = [False,True])
avg_details = avg_details.reset_index(drop=True)
avg_details
```

#<center>RUNNING AVERAGE OF A PLAYER <center/>

```
batsman_ka_runs =  merged_2020_data.groupby('batsman', as_index = False)['batsman_runs'].sum()
batsman_ka_runs.head()

batsman_ka_outs = merged_2020_data[merged_2020_data['dismissal_kind'] != 'retired hurt']
batsman_ka_outs = batsman_ka_outs.groupby('player_dismissed', as_index=False).size()
batsman_ka_outs.columns =['batsman', 'out_data']
batsman_ka_outs

df1 = merged_2020_data[['id','batsman']]
df1

df2 = merged_2020_data[['id','non_striker']]
df2.columns = ['id', 'batsman']
df2

batsman_ka_innings =  pd.concat([df1,df2])
batsman_ka_innings =batsman_ka_innings.groupby('batsman', as_index=False)['id'].nunique()
batsman_ka_innings.columns = ['batsman', 'no_of_innings']
batsman_ka_innings


batsman_avg = batsman_ka_runs.merge(batsman_ka_outs, on='batsman')
batsman_avg['avg_data'] = batsman_avg['batsman_runs'] / batsman_avg['out_data']
batsman_avg = batsman_avg.merge(batsman_ka_innings, on='batsman')


batsman_avg['not_out'] = batsman_avg['no_of_innings'] - batsman_avg['out_data']
batsman_avg = batsman_avg[batsman_avg['no_of_innings'] >5 ].sort_values('avg_data', ascending =False)[['batsman','no_of_innings','batsman_runs','not_out','avg_data']]
batsman_avg
```

```
running_avg = merged_2020_data.copy()
running_avg['id'] = running_avg['id'].astype(str)
running_avg['unique_id_batsman'] = running_avg['id'] + running_avg['batsman']
running_avg.head()
running_avg.sort_values(['date','inning','over','ball']).head(15)
each_day_runs =running_avg.groupby(['date','batsman'],
                                   as_index=False)['batsman_runs'].sum()


filtered_data = running_avg[(~running_avg['dismissal_kind'].isnull()) &
                            (running_avg['dismissal_kind'] != 'retired hurt')]
filtered_data


def check_wicket(row):
  df_date = row['date']
  df_batsman = row['batsman']
  count = filtered_data[(filtered_data['date'] == df_date) &
                        (filtered_data['player_dismissed'] == df_batsman)]

  if len(count) >0:
    return 1
  else:
    return 0


each_day_runs['out_in_match'] = each_day_runs.apply(check_wicket, axis=1)
each_day_runs
```

##Batsman score in the 5 consecutive matches 

```
from tqdm import tqdm

def match5_avg(batsman_name):  
  new_df = pd.DataFrame()
  total_match = len(each_day_runs[each_day_runs['batsman'] == batsman_name])
  if total_match <5:
    bat_rec = each_day_runs[each_day_runs['batsman'] == batsman_name]
    bat_rec['date'] = bat_rec['date'].astype('str')
    out_sum = bat_rec['out_in_match'].sum()
    bat_runs = bat_rec['batsman_runs'].sum()
    if out_sum == 0:
      bat_avg = bat_runs
    else:
      bat_avg = bat_runs/ out_sum
    new_dict = {'date':f"{bat_rec.iloc[0,0]}  to    {bat_rec.iloc[total_match-1,0]}" , 'batsman': batsman_name, 'bat_avg':bat_avg}
    new_df = new_df.append(new_dict,ignore_index=True,sort=False)
  else: 
    for i in range(0,total_match-4):
        bat_rec = each_day_runs[each_day_runs['batsman'] == batsman_name][i:i+5]
        bat_rec['date'] = bat_rec['date'].astype('str')
        out_sum = bat_rec['out_in_match'].sum()
        
        bat_runs = bat_rec['batsman_runs'].sum()
        if out_sum == 0:
          bat_avg = bat_runs
        else:
          bat_avg = bat_runs/ out_sum

        #new_df = pd.DataFrame({'date':[f"{bat_rec.iloc[0,0]}  to    {bat_rec.iloc[4,0]}"] , 'batsman': [batsman_name], 'bat_avg':[bat_avg]})
        new_dict = {'date':f"{bat_rec.iloc[0,0]}  to    {bat_rec.iloc[4,0]}" , 'batsman': batsman_name, 'bat_avg':bat_avg}
        new_df = new_df.append(new_dict,ignore_index=True,sort=False)
        #new_df.append(new_dict,ignore_index=True,sort=False)
  return new_df[['date','batsman','bat_avg']]    

new_df_avg = pd.DataFrame()
all_batsmans = list(each_day_runs['batsman'].unique())
for each  in tqdm(all_batsmans):
  new_df_avg = new_df_avg.append(match5_avg(each))
  
new_df_avg
# result = match5_avg('S Dhawan').append(match5_avg('RR Pant'))
# result
```
